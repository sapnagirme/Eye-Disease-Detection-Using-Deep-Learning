## MODEL TESTING (PREDICTIONS)

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
##{'cataract': 0, 'diabetic_retinopathy': 1, 'glaucoma': 2, 'normal': 3}
import numpy as np

def model_testings(model, img_path):

    img=image.load_img(img_path,target_size=(64,64))

    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    pred=model.predict(x)

    predicted_index=np.argmax(pred)

    disease_labels = {
        0: "Cataract",
        1: "diabetic_retinopathy",
        2: "Glaucoma",
        3: 'Normal'
    }

    if predicted_index in disease_labels:
        predicted_disease = disease_labels[predicted_index]
    else:
        predicted_disease = "Unknown"

    plt.imshow(img)
    plt.title("Input Image")
    plt.axis('off')
    plt.show()


    print("Image passed was: ", img_path)
    print("Predicted Disease Index:", predicted_index)
    print("Predicted Disease:", predicted_disease)
```

```python
model = load_model('eye_cnn.h5')
img_path = '/content/Normal2.jpg'
model_testings(model, img_path)
```
